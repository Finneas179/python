```python
import  matplotlib.pyplot as plt
import  pandas as  pd
import  numpy as np
```

```python
fig,ax=plt.subplots()
ax.plot([1,2,6,4,5],[6,7,8,9,10],marker='o')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot([1,2,3,4,5],[1,4,9,12,25],alpha=1)
plt.show()
```

```python
x=np.arange(0,10,0.01)
y1=np.sin(x)
y2=np.cos(x)
fig,ax=plt.subplots()
#设置透明度
ax.plot(x,y1,marker='*',alpha=0.9,linewidth=10,markersize=10)
ax.plot(x,y2)
plt.show()
```
