```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import title, yticks
```

```python
fig,ax=plt.subplots()
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
x=list(range(1,8))
wuhan=[15,20,22,23,20,18,16]
hei=[-10,-8,-12,-10,-8,-6,1]
ax.plot(x,wuhan,color='r',marker='o',alpha=1,markersize=10,linestyle='-',label='武汉')
ax.plot(x,hei,color='b',linestyle='-',marker='o',markersize=10,label='黑龙江')
plt.title('比较',loc='left',fontsize=20)
plt.xlabel('日期',fontsize=8)
plt.ylabel('温度',fontsize=8)
plt.legend(loc=4)
xticks=['周一','周二','周三','周四','周五','周六','周七']
plt.xticks(x,xticks,rotation=30)
#x原来,xticks修改,rotation倾斜的角度
yticks=list(range(-15,40,5))
plt.yticks(yticks)
for i,j in zip(x,wuhan):
    print(i,j)
    plt.text(i,j,f'{j}C°',horizontalalignment='center')
print('==========')
for i,j in zip(x,hei):
    print(i,j)
    plt.text(i,j,f'{j}C°',horizontalalignment='center')
```
