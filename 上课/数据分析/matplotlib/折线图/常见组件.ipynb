```python
import matplotlib.pyplot as plt
import pandas as pd
import  numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig,ax = plt.subplots()
ax.plot(["a","b","c","d","e","f","g","h"],[1,2,3,4,5,6,7,8])
ax.set_title('test',fontdict={'size':20},loc='center')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.set_xticks([0,1,2,3,4,5,6,7])
ax.set_yticks([0,10,20,30,40,50,60])
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(['1','2','3'],[1,3,5],label='2023')
ax.plot(['1','2','3'],[2,4,5],label='2022')
ax.legend()
plt.show()
```
